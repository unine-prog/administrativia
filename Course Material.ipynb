## What to do now?

You now have a local copy of the course material locally in your Noto personal account. 

Git is a versioning system. If the professor changes something on the notebook in GitHub, you can simply "pull" the latest changes locally from the Github repository. This way, you always have the last copy available. To do this, go to the GitHub panel by clicking the little GitHub icon  on the left, then click the little download cloud icon  on the top right of the panel.

<span style="color:red">WARNING:</span> If you modify your local copy, you won't be able to pull the latest changes as it will cause conflicts between your modifications and mine. If you want to take notes directly in the notebooks, I recommend saving a copy.

## Virtual Environment

To run the Jupyter notebooks, you need a few Python modules that are not installed by default on Noto. The clean and safe way to have a working Python setup that suits your needs is to create a virtual environment and associated kernel. This is the only way that will work on Noto. The complete documentation on how to do this is available in the [Documentation](../Documentation) folder, in [11 Tutorial Envs](../Documentation/11_Tutorial_Envs.ipynb).

**In short...**

Run the following cell (Select the cell and click the little 'run' button at the top or hit Shift+Enter. The Kernel must be Bash on the top right!). It will create a new virtual environment `programming` and an associated Jupyter kernel `Programming`:

```bash
my_venvs_create programming
my_venvs_activate programming
pip install yfinance eurostatapiclient
my_kernels_create programming "Programming"
my_venvs_deactivate
```

You should now be able to run the first Jupyter notebook [01 Introduction](01%20Introduction.ipynb) without any error. Don't forget to select to correct kernel for your project!

## Other platforms

Of course, you can clone the course material GitHub repository to another platform or locally on your computer. To do this, you have to copy the Deploy Key to the appropriate location depending on the system, and use a Git tool or plugin.

* [Git command](https://git-scm.com/downloads) (Windows, Linux, MacOS)
* [Git for Visual Studio Code](https://code.visualstudio.com/docs/sourcecontrol/overview)
* [Git extension for JupyterLab](https://github.com/jupyterlab/jupyterlab-git)
